```python
from gtts import gTTS
import os
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Download necessary NLTK data
nltk.download('punkt')
nltk.download('stopwords')

# Function to clean and preprocess text
def preprocess_text(text):
    # Convert text to lowercase
    text = text.lower()

    # Remove numbers
    text = re.sub(r'\d+', '', text)

    # Remove punctuation
    text = re.sub(r'[^\w\s]', '', text)

    # Tokenize the text
    words = word_tokenize(text)

    # Remove stop words
    stop_words = set(stopwords.words('english'))
    filtered_words = [word for word in words if word not in stop_words]

    # Reconstruct the sentence
    cleaned_text = ' '.join(filtered_words)
    return cleaned_text

# Function to read text from a file
def read_text_from_file(file_path):
    with open(file_path, 'r') as file:
        text = file.read()
    return text

# Get the file path from the user
# file_path = input("Enter the path to the .txt file: ")
file_path = "input_text.txt"
# Read text from the file
try:
    user_input = read_text_from_file(file_path)

    # Preprocess the text
    cleaned_text = preprocess_text(user_input)

    # Show the original and cleaned text
    print(f"\nOriginal Text: {user_input}")
    print(f"Cleaned Text: {cleaned_text}\n")

    # Convert cleaned text to audio
    language = 'en'
    speech = gTTS(text=cleaned_text, lang=language, slow=False)

    # Save the audio file
    output_filename = "output1.mp3"
    speech.save(output_filename)

    # Play the audio file
    os.system(f"start {output_filename}")  # Use 'afplay' for MacOS or 'xdg-open' for Linux

    print(f"The audio has been saved as {output_filename} and is now playing.")

except FileNotFoundError:
    print(f"File not found: {file_path}")
```
